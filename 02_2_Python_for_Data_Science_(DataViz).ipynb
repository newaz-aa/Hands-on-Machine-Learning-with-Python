```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data_file_lab2.csv')
data
```

```python
data.head()
```

# Bar Plot

```python
sns.barplot(x=data.index, y='SMOTE', data=data);
```

```python
sns.barplot(x=data.index, y='SMOTE', data=data);
plt.xlabel('Datasets')
plt.ylabel('Score')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=data.index, y='SMOTE', data=data);
plt.xlabel('Datasets')
plt.ylabel('Score')
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x=data.index, y='SMOTE', data=data, palette = "cividis");
plt.xlabel('Datasets')
plt.ylabel('Score')
```

# Categorical Plots

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.catplot(data=df, x="age", y="class")
```

```python
sns.catplot(data=df, x="age", y="class", kind="box")
```

# Count plot

```python
sns.countplot(x="deck",
data=df,
palette="Greens_d")
```

```python
sns.countplot(x="embarked",
data=df,
palette="Greens_d")
```

# Box plot

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=df)
```

# Violin plot

```python
sns.violinplot(x="age", y="sex", hue="survived", data=df)
```
