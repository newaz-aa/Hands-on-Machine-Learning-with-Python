# Titanic Dataset

```python
df = sns.load_dataset("titanic")
df
```

# Scatter Plot

```python
plt.scatter(df['age'], df['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

# Histogram

```python
plt.hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

# Sub Plot

```python
plt.figure(figsize=(10,5))

# First subplot: Age distribution
plt.subplot(1, 2, 1)
plt.hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')

# Second subplot: Fare distribution
plt.subplot(1, 2, 2)
plt.hist(df['fare'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Fare Distribution')

plt.tight_layout()
plt.show()
```

# Categorical Plots

```python
sns.catplot(data=df, x="age", y="class")
```

```python
sns.catplot(data=df, x="age", y="class", kind="box")
```

# Count plot

```python
sns.countplot(x="deck", data=df, palette="Greens_d")
```

```python
sns.countplot(x="embarked", data=df, palette="Greens_d")
```


# Pie Chart

```python
df['class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'], startangle=90)
plt.title('Class Distribution')
plt.ylabel('')  # Hides the y-label
plt.show()
```

# Box plot

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=df)
```

# Violin plot

```python
sns.violinplot(x="age", y="sex", hue="survived", data=df)
```

#Pair plot

```python
sns.pairplot(df, hue="survived")
```

```python
sns.pairplot(df, hue="survived", kind = "hist",
            x_vars=["age", "pclass", "fare"],
            y_vars=["age", "pclass", "fare"])
```

#Heat map

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

#Joint Plot

```python
sns.jointplot(x='age', y='fare', data=df, kind='hex')
```

#Swarm Plot



```python
sns.swarmplot(x='sex', y='age', data=df)
```

#Regression Analysis using regplot and lmplot

```python
sns.regplot(x='age', y='fare', data=df,scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Regression Plot of Age vs Fare')
plt.show()
```

```python
sns.lmplot(x="age", y="fare", data=df, hue="survived", aspect=1.5, ci=None)
```
